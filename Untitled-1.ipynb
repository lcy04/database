```python
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
from tkinter import *

def home():
     new_window = tk.Toplevel(root)
     new_window.title("Home Window")
     sql_query = 'SELECT 姓名,專業,辦公室 FROM 醫生表'
     cursor.execute(sql_query)
     

     tree = ttk.Treeview(new_window, columns=("姓名", "專業", "辦公室"))
     tree.heading("#0", text="姓名")
     tree.heading("#1", text="專業")
     tree.heading("#2", text="辦公室")

     for row in cursor.fetchall():
          tree.insert('', 'end',text=row['姓名'],values=(row['專業'], row['辦公室']))
    

     tree.pack()
     root.mainloop()



def work():
     new_window = tk.Toplevel(root)
     new_window.title("Home Window")
     sql_query = 'SELECT 科室號碼,醫生 FROM 看診表'
     cursor.execute(sql_query)
     

     tree = ttk.Treeview(new_window, columns=("科室號碼", "醫生"))
     tree.heading("#0", text="科室號碼")
     tree.heading("#1", text="醫生")

     for row in cursor.fetchall():
          tree.insert('', 'end',text=row['科室號碼'],values=(row['醫生']))
     

     tree.pack()
     root.mainloop()

def people():
     pwindow = tk.Toplevel(root)
     pwindow.title("Home Window")
     sql_query = 'SELECT 編號,姓名,科室號碼 FROM 病人'
     cursor.execute(sql_query)
     

     tree = ttk.Treeview(pwindow, columns=("編號", "姓名", "科室號碼"))
     tree.heading("#0", text="編號")
     tree.heading("#1", text="姓名")
     tree.heading("#2", text="科室號號")

     for row in cursor.fetchall():
          tree.insert('', 'end',text=row['編號'],values=(row['姓名'], row['科室號碼']))
     
     tree.pack()
     root.mainloop()
  


```

```python
def add():
    # 創建子視窗
    pwindow = tk.Toplevel(root)
    pwindow.title("新增病人")

    tk.Label(pwindow, text='編號：').pack()
    number_name = tk.Entry(pwindow)
    number_name.pack()

    tk.Label(pwindow, text='姓名：').pack()
    entry_name = tk.Entry(pwindow)
    entry_name.pack()

    tk.Label(pwindow, text='病歷號碼：').pack()
    entry_medical_record_number = tk.Entry(pwindow)
    entry_medical_record_number.pack()

    def submit():
        try:
            # 取得使用者輸入的資料
            number_val = number_name.get()
            patient_name_val = entry_name.get()
            medical_record_number_val = entry_medical_record_number.get()

            

            # 執行 SQL 查詢，插入新的病人資訊
            sql_query = "INSERT INTO 病人 (編號, 姓名, 科室號碼) VALUES (?, ?, ?)"

            cursor.execute(sql_query, (number_val, patient_name_val, medical_record_number_val))
            
            # 提交變更並關閉連接
            conn.commit()
           

            # 提示使用者新增成功
            tk.messagebox.showinfo('新增病人', '病人新增成功！')

            # 關閉子視窗
            pwindow.destroy()
        except Exception as e:
        
            tk.messagebox.showerror('錯誤', f'發生錯誤：{e}')

    def de():
        try:
            # 取得使用者輸入的資料
            number_val = number_name.get()

            # 執行 SQL 查詢，刪除符合條件的病人資訊
            sql_query = "DELETE FROM 病人 WHERE 編號 = ?"
            cursor.execute(sql_query, (number_val,))

            
            # 提交變更並關閉連接
            conn.commit()
           

            tk.messagebox.showinfo('刪除病人', '病人刪除成功！')

            
            pwindow.destroy()
        except Exception as e:
        
            tk.messagebox.showerror('錯誤', f'發生錯誤：{e}')



    # 創建提交按鈕
    button_submit = tk.Button(pwindow, text='提交', command=submit)
    button_del=tk.Button(pwindow,text='刪除',command=de)
    button_submit.pack()
    button_del.pack()
    
```

```python
def query_data_by_patient_id():
     
    pwindow = tk.Toplevel(root)
    pwindow.title("病人查詢系統")

    tk.Label(pwindow, text='編號：').pack()
    number_name = tk.Entry(pwindow)
    number_name.pack()
    tree = ttk.Treeview(pwindow, columns=("編號", "病人姓名", "科室號碼", "醫生姓名", "專業", "辦公室"))
    tree.heading("#1", text="編號")
    tree.heading("#2", text="病人姓名")
    tree.heading("#3", text="科室號碼")
    tree.heading("#4", text="醫生姓名")
    tree.heading("#5", text="專業")
    tree.heading("#6", text="辦公室")
    def search():
        try:
            patient_id = number_name.get()

            sql_query = f"""
                SELECT 病人.編號, 病人.姓名 AS 病人姓名, 科室.科室號碼, 科室.醫生 AS 醫生姓名, 醫生.專業, 醫生.辦公室
                FROM 病人
                INNER JOIN 看診表 AS 科室 ON 病人.科室號碼 = 科室.科室號碼
                INNER JOIN 醫生表 AS 醫生 ON 科室.醫生 = 醫生.姓名
                WHERE 病人.編號 = ?;
            """
            cursor.execute(sql_query, (patient_id,))

            # 取得查詢結果
            results = cursor.fetchall()
            for row in results:
                tree.insert('', 'end', values=row)
            '''
            L1=tk.Label(root,text=results,bg='#CCCBCB',fg='#474242',
            font=("San Francisco",18,"bold"),cursor='heart')
            L1.pack()
            '''
        except Exception as e:
            tk.messagebox.showerror('錯誤', f'發生錯誤：{e}')


    button_query_by_patient_id = tk.Button(pwindow, text='查詢', command=search)
    button_query_by_patient_id.pack()
    tree.pack()
```

```python
def update_patient():
    # Create a new window for updating patient information
    update_window = tk.Toplevel(root)
    update_window.title("修改病人資料")

    # Labels and Entry widgets for the update window
    tk.Label(update_window, text='編號：').grid(row=0, column=0)
    number_name = tk.Entry(update_window)
    number_name.grid(row=0, column=1)

    tk.Label(update_window, text='新姓名：').grid(row=1, column=0)
    new_name_entry = tk.Entry(update_window)
    new_name_entry.grid(row=1, column=1)

    tk.Label(update_window, text='新病歷號碼：').grid(row=2, column=0)
    new_medical_record_entry = tk.Entry(update_window)
    new_medical_record_entry.grid(row=2, column=1)

    
    def perform_update():
        try:
            # Get values from entry widgets
            patient_id = number_name.get()
            new_name = new_name_entry.get()
            new_medical_record = new_medical_record_entry.get()

            # Execute SQL update query
            update_query = f"""
                UPDATE 病人
                SET 姓名 = ?, 科室號碼 = ?
                WHERE 編號 = ?;
            """
            cursor.execute(update_query, (new_name, new_medical_record, patient_id))

            
            conn.commit()

            
            tk.messagebox.showinfo('成功', '病人資料更新成功！')

            
            update_window.destroy()

        except Exception as e:
            tk.messagebox.showerror('錯誤', f'發生錯誤：{e}')
    update_button = tk.Button(update_window, text='更新', command=perform_update)
    update_button.grid(row=3, column=0, columnspan=2)

     
```

```python
import tkinter as tk
root = tk.Tk()
root.title('主畫面')
root.configure(bg='#CCCBCB') 
root.geometry("400x300+200+300")

frameUpper=tk.Frame (root, bg="#FFC1E0")#上半部分
frameUpper.pack()

L1=tk.Label(frameUpper,text='主控台',bg='#ff6363',fg='#474242',
            font=("Viner Hand ITC",18,"bold"),cursor='heart')
L1.pack()
B1=tk.Button(frameUpper,text='病人資訊',relief="ridge",
            activebackground='#BE77FF',#設定滑鼠位於按鈕時的背景顏色
            activeforeground='#FFFFFF',#設定滑鼠位於按鈕時的前景顏色
             command=people,#設定按鈕的狀態
             cursor='heart')
B2=tk.Button(frameUpper,text='看診醫生',relief="ridge",
            activebackground='#BE77FF',#設定滑鼠位於按鈕時的背景顏色
            activeforeground='#FFFFFF',#設定滑鼠位於按鈕時的前景顏色
             command=work,#設定按鈕的狀態
             cursor='heart')
    
button = tk.Button(frameUpper,text='醫生清單',relief="ridge",
            activebackground='#BE77FF',#設定滑鼠位於按鈕時的背景顏色
            activeforeground='#FFFFFF',#設定滑鼠位於按鈕時的前景顏色
             command=home,#設定按鈕的狀態
             cursor='heart')
button.pack(side=tk.LEFT)
B1.pack(side=tk.LEFT)
B2.pack(side=tk.LEFT)
framedown=tk.Frame (root, bg="#2db88a")#上半部分
framedown.pack()

button2 = tk.Button(framedown,text='新增/刪除病人',relief="ridge",
            activebackground='#BE77FF',#設定滑鼠位於按鈕時的背景顏色
            activeforeground='#FFFFFF',#設定滑鼠位於按鈕時的前景顏色
             command=add,#設定按鈕的狀態
             cursor='heart')
             
button2.pack(side=tk.RIGHT)

button3 = tk.Button(root,text='病人查詢系統',relief="ridge",
            activebackground='#BE77FF',#設定滑鼠位於按鈕時的背景顏色
            activeforeground='#FFFFFF',#設定滑鼠位於按鈕時的前景顏色
             command=query_data_by_patient_id,#設定按鈕的狀態
             cursor='heart')
             
button3.pack()

update_button = tk.Button(framedown, text='修改病人資料', command=update_patient)
update_button.pack(side=tk.RIGHT)
root.mainloop()

```
